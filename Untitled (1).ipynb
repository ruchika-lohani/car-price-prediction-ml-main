```python
X=df[['insurance_validity','fuel_type','kms_driven','ownsership','transmission']]
y=df['price(in lakhs)']
```

```python

from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
```

```python
# Fix missing values
X = X.fillna(X.mean(numeric_only=True))
X = X.fillna(0)

models = {
    'LinReg': LinearRegression(),
    'KNN-3': KNeighborsRegressor(n_neighbors=3),
    'SVM-Poly': SVR(kernel='poly')
}

for name, model in models.items():
    model.fit(X, y)
    sc = model.score(X, y)
    print(name, sc)
```

```python
final_model = models['KNN-3']
```

```python
d1={'Comprehensive': 0,'Third Party insurance': 1,'Zero Dep': 2,'Not Available': 3,'Third Party': 1}
d2={'Petrol': 0, 'Diesel': 1, 'CNG': 2}
d3={'Manual': 0, 'Automatic': 1}
d4={'First Owner': 1,'Second Owner': 2,'Third Owner': 3,'Fifth Owner': 5,'Fourth Owner': 4}
```

```python

insurance_validity = input('insurance_validity:')
fuel_type=input('fuel_type:')
kms_driven =input ('kms_driven:')
ownsership = input('ownsership:')
transmission =input ('transmission:')

insurance_validity =df1[insurance_validity]
fuel_type =df2[fuel_type]
kms_driven = int(kms_driven)
ownsership =df3[ownsership]
transmission =df4[transmission]
test= [[insurance_validity,fuel_type,ownsership,kms_driven,transmission]]
yp=int(final_model.predict(test)[0])
print('Predicted car price in {} RS'.format(yp))
```
